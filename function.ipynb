```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
    
def cal(a,b,o):
    if o=='+':
        add(a,b)
    elif o=='-':
        sub(a,b)
    elif o=='*':
        mul(a,b)
    elif o=='/':
        div(a,b)
    else:
        print("Invalid operator")
    
a = int(input("Enter a : "))
b = int(input("Enter b : "))
o = input("Enter + - * / ")
cal(a,b,o)
```

```python
def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def mul(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
    
def cal(a,b,o):
    if o=='+':
        print(add(a,b))
    elif o=='-':
        print(sub(a,b))
    elif o=='*':
        print(mul(a,b))
    elif o=='/':
        print(div(a,b))
    else:
        print("Invalid operator")
    
a = int(input("Enter a : "))
b = int(input("Enter b : "))
o = input("Enter + - * / ")
cal(a,b,o)
```
